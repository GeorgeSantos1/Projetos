```python
import pandas as pd
import yfinance as yf
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import matplotlib.style as style
import plotly.express as px

style.use('fivethirtyeight')
```

```python
# Obtendo Dados de Preços de Ações da Magazine Luiza SA
ticket = yf.Ticker('MGLU3.SA')

# Definindo Período
df = ticket.history(interval='1d',start='2016-07-01',end='2021-07-02',prepost=True)
```

```python
df['Date'] = df.index

plt.figure(figsize=[12,6])
plot = sns.lineplot(data=df,x='Date',y='Close',color=[204/255,121/255,167/255])
plt.suptitle('Ações Magazine Luiza (MGLU3)',size=19,weight = 'black')
plt.title('Julho/2017 - Julho/2021',size=17)
plt.ylabel('Valor de Fechamento (R$)') , plt.xlabel('')
plt.setp(plot.get_xticklabels(), rotation=-30);
```

Com base na série histórica pode-se afirmar que as ações da Magazine Luiza valorizaram de forma considerável nos últimos 5 anos, principalmente durante a pandemia de COVID-19.

#### **Decomposição de séries temporais**

```python
decomposição = seasonal_decompose(df[['Close']],extrapolate_trend=30,freq=30)
```

```python
# Função que gera gráfico para cada um dos elementos
def graf_decomp(data,title,color = [86/255,180/255,223/255]):
  plot = sns.lineplot(data=data.reset_index(),x='Date',y='Close',color = color);
  plt.title(title,size=14,weight = 'bold')
  plt.ylabel(''), plt.xlabel('')
  plt.setp(plot.get_xticklabels(), rotation=-30);
```

```python
# Gráfico da série Observada
plt.figure(figsize=[12,12])
plt.subplot(411)
graf_decomp(data=decomposição.observed,title='Observado')

# Gráfico de tendência
plt.subplot(412)
graf_decomp(data=decomposição.trend,title='Tendência')

# Gráfico de sazonalidade
plt.subplot(413)
graf_decomp(data=decomposição.seasonal,title='Sazonalidade')

# Gráfico de Resíduos
plt.subplot(414)
graf_decomp(data=decomposição.resid,title='Resíduo',color=[213/255,94/255,0])
plt.tight_layout()
```

#### **Média Móveis**

```python
df_ma = pd.DataFrame({'Observado': df['Close'],
                      'MA_7dias': df['Close'].rolling(7).mean(),
                      'MA_14dias': df['Close'].rolling(14).mean(),
                      'MA_21dias': df['Close'].rolling(21).mean(),
                      'MA_30dias': df['Close'].rolling(30).mean()})
```

```python
df_ma = pd.melt(df_ma.reset_index(),id_vars=['Date'],
                value_vars=['Observado','MA_7dias', 'MA_14dias','MA_21dias', 'MA_30dias'],
                var_name='TS',value_name='Close')
```

```python
def graf_ma(ts):
  plot = sns.lineplot(data= df_ma[(df_ma.TS == 'Observado') | (df_ma.TS == ts)],x='Date',y='Close',hue='TS',alpha = 0.5)
  plt.ylabel(''), plt.xlabel('')
  plt.setp(plot.get_xticklabels(), rotation=-30);
```

```python
plt.figure(figsize=[10,6])
plt.subplot(221)
graf_ma('MA_7dias')
plt.subplot(222)
graf_ma('MA_14dias')
plt.subplot(223)
graf_ma('MA_21dias')
plt.subplot(224)
graf_ma('MA_30dias');
```
